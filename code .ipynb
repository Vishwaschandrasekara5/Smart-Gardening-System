## Task 1: Import Required Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import load_iris 
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import KMeans, AgglomerativeClustering 
from scipy.cluster.hierarchy import linkage, dendrogram 
```

## Task 2: Feature Scaling and Selection

```python
# Load Iris dataset 
iris = load_iris() 
X = pd.DataFrame(iris.data, columns=iris.feature_names) 

# Normalize features 
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(X) 
```

## Task 3: Apply K-Means Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=0) 
kmeans.fit(X_scaled) 
labels_kmeans = kmeans.labels_
```

## Task 4: Visualize K-Means Clustering 

```python
plt.figure(figsize=(8,6)) 
plt.scatter(X_scaled[:, 2], X_scaled[:, 3], 
c=labels_kmeans, cmap='viridis', s=60) 
centers = kmeans.cluster_centers_ 
plt.scatter(centers[:, 2], centers[:, 3], c='black', s=200, 
alpha=0.75, marker='X') 
plt.xlabel("Petal Length (normalized)") 
plt.ylabel("Petal Width (normalized)") 
plt.title("K-Means Clustering (Iris)") 
plt.show() 
```
